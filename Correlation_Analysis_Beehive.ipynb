
# Correlation Analysis (Beehive Acoustic Features)

## Aim
This notebook combines extracted features (RMS, MFCC summaries, spectral features, ZCR) and environmental variables (temperature, humidity) to compute and visualise correlations.

The goal is to identify relationships between acoustic behaviour of the hive and environmental conditions.



## Notes
- Use aligned CSVs produced in extraction stage.
- Correlation does **not imply causation**.
- Because signals were normalised earlier, interpretation focuses on **relative trends**.


```python

import os
from glob import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Configuration

```python

FEATURE_FOLDER = "data/features"  # adjust if needed
OUTPUT_FOLDER = "data/analysis"

os.makedirs(OUTPUT_FOLDER, exist_ok=True)
```

## Load feature datasets

```python

# Load all feature CSVs (assuming consistent timestamp column)
files = glob(os.path.join(FEATURE_FOLDER, "*.csv"))

dfs = []
for f in files:
    df = pd.read_csv(f, parse_dates=["timestamp"])
    dfs.append(df)

# Merge all on timestamp (inner join for consistency)
from functools import reduce

merged = reduce(lambda left, right: pd.merge(left, right, on="timestamp", how="inner"), dfs)

merged.head()
```

## Clean data

```python

merged = merged.dropna()

# Optional: remove non-numeric columns
numeric_df = merged.select_dtypes(include=[np.number])

numeric_df.head()
```

## Correlation matrix

```python

corr = numeric_df.corr()

corr
```

## Heatmap

```python

plt.figure(figsize=(10,8))
plt.imshow(corr, aspect='auto')
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```


## Interpretation

Correlation values range from -1 to 1:

- **+1** → strong positive relationship
- **0** → no linear relationship
- **-1** → strong negative relationship

In this project, useful insights may include:
- correlation between RMS and temperature → activity vs environment
- spectral features vs humidity → acoustic response to conditions
- relationships between features (e.g. centroid vs bandwidth)

These relationships will guide further analysis (PCA, clustering).
